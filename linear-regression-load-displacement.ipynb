# Wing Crack Growth using Linear Regression at pre-existing crack inclination angle β=45°

Dataset: wingcrackgrowth.csv

Independant variable X: Load (kN) 
Dependant variable Y: Displacement (mm)

Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Read the dataset

```python
WingCrack = pd.read_csv("wingcrackgrowth.csv")
WingCrack.head(12)
```

Visualize the dataset

```python
sns.jointplot(x='Load (kN)', y='Displacement (mm)', data = WingCrack)
sns.pairplot(WingCrack)
sns.lmplot(x='Load (kN)', y='Displacement (mm)', data = WingCrack)
```

Create Train and Test dataset

```python
y = WingCrack['Displacement (mm)']
X = WingCrack[['Load (kN)']]

from sklearn.model_selection import train_test_split
```

Split the dataset into train data and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

Shape of train data

```python
X_train.shape
```

Import Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

Train the model 

Equation of line y = mx + b

```python
regressor = LinearRegression(fit_intercept =True)
regressor.fit(X_train,y_train)

print('Linear Model Coefficient (m): ', regressor.coef_)
print('Linear Model Coefficient (b): ', regressor.intercept_)
```

Test the Model

```python
y_predict = regressor.predict( X_test)
y_predict
y_test
```

Visualize train set results

Blue line indicates the line of best fit

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.ylabel('Displacement [mm]')
plt.xlabel('Load [kN]')
plt.title('Wing Crack Growth Generated vs. Load @Wing Crack(Training dataset)')
```

Visualize test set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.ylabel('Displacement [mm]')
plt.xlabel('Load [kN]')
plt.title('Wing Crack Growth Generated vs. Load @Wing Crack(Test dataset)')
```
